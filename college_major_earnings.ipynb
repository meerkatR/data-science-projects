# Visualizing Earnings Based on College Majors

## Import and explore data

```python
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
recent_grads = pd.read_csv("recent-grads.csv")
recent_grads.iloc[0]
```

```python
recent_grads.head()
```

```python
recent_grads.tail()
```

```python
recent_grads.describe()
```

## Remove Missing

```python
raw_data_count = len(recent_grads)
recent_grads = recent_grads.dropna()
cleaned_data_count = len(recent_grads)
print(raw_data_count, cleaned_data_count)
```

## Explore relationships between columns

```python
recent_grads.plot(x = 'Sample_size', y = 'Median', kind = 'scatter', title = 'Median Income vs. Sample Size')
```

```python
recent_grads.plot(x = 'Sample_size', y = 'Unemployment_rate', kind = 'scatter', title = 'Unemployment Rate vs. Sample Size')
```

```python
recent_grads.plot(x = 'Full_time', y = 'Median', kind = 'scatter', title = 'Full Time Employment vs. Median Income')
```

```python
recent_grads.plot(x = 'ShareWomen', y = 'Unemployment_rate', kind = 'scatter', title = '% of women vs. Unemployment Rate')
```

```python
recent_grads.plot(x = 'Men', y = 'Median', kind = 'scatter', title = '# of Men vs. Median Income')
```

```python
recent_grads.plot(x = 'Women', y = 'Median', kind = 'scatter', title = '# of Women vs. Median Income')
```

Do students in more popular majors make more money? Not necesarily. Some very popular majors with more than 100,000 people did not make any more money compared to others.  While some less popular majors can earn more.

```python
recent_grads.plot(x = 'Total', y = 'Median', kind = 'scatter', title = 'Median Income vs. Total # of people')
```

Do students that majored in subjects that were majority female make more money?

```python
recent_grads.plot(x = 'ShareWomen', y = 'Median', kind = 'scatter', title = 'Median Income vs. % of Women')
```

There is a weak negative correlation between % of women and median income. Students that majored in subjects that were majority female tend to make less money.

Is there any link between the number of full-time employees and median salary?  Not really.
